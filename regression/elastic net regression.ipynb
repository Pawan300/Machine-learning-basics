```python
x=np.array([np.ones(train.shape[0]),train[train.columns[0]],train[train.columns[1]],train[train.columns[2]],train[train.columns[3]]]).T
y=train[train.columns[4]]
m=train.count()[0] 
y_actual=test[test.columns[4]]
beta=0.001
```

```python
def regression(epoch):
    theta=np.ones(5)
    err=[]
    for i in range(epoch):
        y_predict=theta.T.dot(x.T)
        err.append(1/(2*m)*(np.sum(np.square(y_predict-y))+beta*(np.sum(theta))+(1-beta)*(np.sum(np.square(theta)))))
        if(err[i]<0.01):
            break
        else:
            theta=gradient(theta,y_predict)
    ploterror(err)
    return(theta)
```

```python
def gradient(theta,y_predict):
    alpha=0.00000001
    theta=theta-alpha/m*(x.T.dot(y_predict-y)+beta*theta.T+(1-beta)*theta.T)
    return(theta)
```

```python
def solution(n):
    theta=regression(n)
    y=np.array([np.ones(test.shape[0]),test[test.columns[0]],test[test.columns[1]],test[test.columns[2]],test[test.columns[3]]]).T
    return(y.dot(theta))
```

```python
def error(n):
    prediction=solution(n)
    z=prediction-y_actual
    return(math.sqrt(np.sum(np.square(z))/m))
```

```python
def ploterror(err):
    print("Plot between error and number of iteration is :")
    plt.plot([i for i in range(len(err))],err)
```
