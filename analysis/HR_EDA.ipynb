# Univariate Analysis

## Categorical Variables

```
def get_unique(data,var):
  print(len(data[var].unique()),"Distinct categories:",data[var].unique())
```

```
'''
for categorical variables univariate analysis
'''
def cat_uni_analysis(var):
  print(len(data[var].unique()),"Distinct categories:",data[var].unique()) #unique categories
  print("No. of missing values:",data[var].isna().sum()) #missing values
  sns.set(style="darkgrid")
  ax = sns.countplot(x=var,data=data) #countplot
  plt.title('Number of employees by '+var)
  plt.ylabel('Number of employees')
  plt.xlabel(var)
  plt.xticks(rotation = 90)
  plt.show();
```

```
'''
for numerical variables univariate analysis
'''
def num_uni_analysis(var):
  print(len(data[var].unique()),"Distinct values:",data[var].unique()) #unique categories
  print("No. of missing values:",data[var].isna().sum()) #missing values
  print(data[var].describe())
  sns.set()
  sns.distplot(data[var],kde = False) #distplot
  plt.title('Histogram of '+var)
  plt.xlabel(var)
  plt.ylabel('Number of employees')
  plt.show();
  sns.set()
  ax1 = sns.boxplot(x=var,data=data) #boxplot for detecting outliers
```

# Bivariate Analysis

```
def num_bi_analysis(var1,var2):
  sns.set()
  ax = sns.scatterplot(x=var1, y=var2, data=data)
  plt.show();
```

```
num_var = ['no_of_trainings', 'age','length_of_service','avg_training_score']
comb = combinations(num_var,2)
```

### Scatter plots

```
for i in comb:
  num_bi_analysis(i[0],i[1])
```

### Correlation

```
corr = data[num_var].corr()
```

```
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
```

```
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Average training score and no of trainings are correlated
